```python
# Gerekli kütüphaneler
from ultralytics import YOLO
import cv2
import pytesseract
import matplotlib.pyplot as plt
import os

# Tesseract yolu (Windows kullananlar için)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR'

# YOLO modelini yükle (modelin yolunu kendi path'in ile değiştir)
model = YOLO('runs/detect/train/weights/best.pt')

# Görselin yolu
img_path = 'arac.jpg'  # Değiştirilebilir

# Görseli oku
img = cv2.imread(img_path)

# Model ile tespit yap
results = model(img_path)

# Tespit edilen plakalar üzerinde OCR uygula
for r in results:
    for box in r.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Kutu koordinatları
        plaka_img = img[y1:y2, x1:x2]           # Kutuya göre görseli kırp

        # OCR işlemi yap
        ocr_text = pytesseract.image_to_string(plaka_img, config='--psm 7')
        ocr_text = ocr_text.strip()

        # Kutucuk çiz ve yazı ekle
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, ocr_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

# Renk dönüşümü (BGR -> RGB) ve gösterim
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("OCR ile Plaka Tanıma")
plt.show()
```
